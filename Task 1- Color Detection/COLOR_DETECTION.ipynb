```python
#importing the requires libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the input images
img=cv2.imread('colours.JPG')
```

```python
#visualising the input image (Default it is in BGR format)
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
#converting the BGR format to RGB format
grid_RGB= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#visualising the RGB format image
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
#converting the RGB format image to HSV format
grid_HSV= cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
#giving the lower hsv and upper hsv values for the color you want to detect(detecting the single color)
lower=np.array([85,150,0])
upper=np.array([95,255,255])
```

```python
#creating the mask for the image
mask= cv2.inRange(grid_HSV,lower,upper)
```

```python
#visualising the masked image
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
#applying the bitwise and operation on the masked image
res= cv2.bitwise_and(grid_RGB,grid_RGB,mask=mask)
```

```python
#visualising the final output image with the color we want to detect
plt.figure(figsize=(20,8))
plt.imshow(res)
```

```python
#giving the lower and upper hsv values for color 1 (detecting multiple colors at once)
lower_1=np.array([95,150,0])
upper_1=np.array([110,255,255])
```

```python
#giving the lower and upper hsv values for color 2 (detecting multiple colors at once)
lower_2=np.array([145,150,0])
upper_2=np.array([155,255,255])
```

```python
#creating a mask for color 1
mask_1= cv2.inRange(grid_HSV,lower_1,upper_1)
```

```python
#creating the mask for color 2
mask_2= cv2.inRange(grid_HSV,lower_2,upper_2)
```

```python
mask=mask_1+mask_2
```

```python
#visualising the masked output
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
#applying the bitwise and operation on masked output
res1= cv2.bitwise_and(grid_RGB,grid_RGB,mask=mask)
```

```python
#visualising the final output with the multiple colors being detected
plt.figure(figsize=(20,8))
plt.imshow(res1)
```
